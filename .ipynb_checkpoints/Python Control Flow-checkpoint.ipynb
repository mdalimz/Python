1. Write a Python program to find those numbers which are divisible by 5 and multiples of 7, between 1100 and 1700 (both included).

```python
n1=[]
for x in range(1500, 2701):
    if (x%7==0):
        n1.append(str(x))
print(','.join(n1))
```

```python
nl=[]
for x in range(1500, 2701):
    if (x%7==0) and (x%5==0):
        nl.append(str(x))
print (','.join(nl))
```
